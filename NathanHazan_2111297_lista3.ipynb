#1) Salve a base de dados escolhida no seu drive e importe para um notebook do Colab, criando um Dataframe

```python
from google.colab import drive
import pandas as pd
drive.mount('/content/drive')
with open('/content/drive/My Drive/german_credit/class_german_credit.csv') as file:
  dados = pd.read_csv(file)
dados = pd.DataFrame (dados)


```

#2) Inspecione as colunas do conjunto de dados (Nome das colunas e tipo de dados)



```python
print(dados.info()) 
```

#3) Escolha uma coluna com valores numéricos e apague todas as linhas que possuam valor nulo


```python
dados["Duration"].dropna()
display(dados)
```

#4) Calcule a média desta mesma coluna


```python
import numpy as np
media = np.mean(dados['Duration'])
print(media)

```

#5) Escolha um critério para eliminar as linhas que possuam valores muito distantes da média e crie um novo Dataframe.


```python
#Meu critério foi: se o for maior a 10% ele sera cortado
filtro  = dados['Duration'] > media/10
dados2 = dados[filtro]
display(dados2)
```

#6) Crie um novo Dataframe que contenha apenas 5 colunas do conjunto anterior.



```python

dados2.drop(columns=['Age','Saving accounts','Credit amount','Risk','Housing'], inplace = True)

display(dados2)
```

#7) Escolha duas colunas numéricas e crie uma nova coluna que siga a regra: (Coluna A - Coluna B)*número aleatório (entre 0 e 1).

```python
import random
dados2.insert(1, 'colunaNova', ((dados2['Duration'] - dados2['Job'])* random.random()).astype(int))

display(dados2)
```

#8) Embaralhe aleatoriamente o dataframe (índices e valores).

```python
dados2 = dados2.reindex(np.random.permutation(dados2.index))

display(dados2)
```

#9) Ordene o Dataframe seguindo um critério definido por você (Explique o critério)

```python
#Meu critério foi: ordenar por duracao comecando na menor
dados2 = dados2.sort_values(by='Duration', ascending=True)
pd.set_option('display.max_columns', 7)
pd.set_option('display.max_rows', 5)
display(dados2)
```

#10) Escolha uma coluna numérica e um coluna com strings e utilize a função Groupby para mostrar os totais da  coluna.

```python
grupo = dados2.groupby('colunaNova')[['colunaNova']].count()
grupo2 = dados2.groupby('Purpose')[['Purpose']].count()
display(grupo)
display(grupo2)
```

#11) Utilize a função MAPS para fazer algum tipo de substituição no Dataframe (explique)


```python
#Criamos uma nova coluna com o nome 'kname' a partir da coluna ' Duration' seguindo as regras do dicionário abaixo

dados2['Purpose'] = dados2['Duration'].map( {1: '4', 2: '60', 3: '72'})


pd.set_option('display.max_columns', 7)
pd.set_option('display.max_rows', 10)
display(dados2)
```
